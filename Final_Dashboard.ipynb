# ข้อมูลชุดที่ 2 : alldb.csv

```python
data5 = os.path.join(path,'alldb.csv') #แล้วก็ชี้ไปที่ไฟล์
print(data5)
```

```python
data6 = pd.read_csv(data5)  #กำหนดตัวแปร data6 เป็นข้อมูล ไฟล์ alldb.csv
data6  #เรียกดูตารางของข้อมูล ไฟล์ alldb.csv
```

## ตัดเอาเฉพาะคอลัมน์ที่ต้องการ

```python
data7= data6[['user','RowKey','date','phone','site','time','urine','ov','PartitionKey','title','age']]
data7
```

## ตรวจดู type ของข้อมูล และดูจำนวน missing ในแต่ละตัวแปร

```python
data7.dtypes    # ดู type ของข้อมูล
```

```python
data7.isnull().sum()     # sum ดูจำนวนของค่า missing
```

## เปลี่ยน Type Data 

```python
import numpy as np
```

### User 
เป็นตัวเลข 5 หลัก (ตัวเลขเท่านั้น)

```python
data7 = data7[(data7['user'].astype(str).str.isdigit())& (data7['user'].str.len() == 5)]     #.str.isdigit ใช้เพื่อตรวจสอบว่าค่าใดที่เป็นตัวเลข # เปลี่ยนคอลัมน์ user ให้มีตัวเลข 5 ตัว 
data7
```

### RowKey 
เป็น String

```python
data7['RowKey'] = data7['RowKey'].astype('str')
data7
```

### date 
เป็น YYYY-MM-DD

```python
data7['date'] = pd.to_datetime(data7['date'])    # format = "%d/%m/%Y" 
data7
```

```python
data7['date'] = data7[['date']].fillna(method="ffill")      # เติมค่าว่างของวันที่ด้วยค่าก่อนหน้าหรือค่าข้างบน
data7
```

### phone 
เป็น String

```python
data7['phone'] = data7['phone'].astype('str')
data7
```

### site 
เป็นตัวเลข 5 หลัก (ตัวเลขเท่านั้น)

```python
data7 = data7 [(data7 ['site'].astype(str).str.isdigit())& (data7['user'].str.len() == 5)]     #.str.isdigit ใช้เพื่อตรวจสอบว่าค่าใดที่เป็นตัวเลข # เปลี่ยนคอลัมน์ user ให้มีตัวเลข 5 ตัว 
data7 
```

### time 
เป็น hh-mm-ss

```python
data7['time'] = data7['time'].replace(':','-',regex=True) # กำหนดรูปแบบของ time เป็น hh-mm-ss
data7
```

### urine 
เป็น String

```python
data7['urine'] = data7['urine'].astype('str')
data7
```

### ov 
เป็นตัวเลข มีค่าตั้งแต่ 0-4 (0 คือไม่พบโรค 1-4 คือพบโรคที่ระดับความรุนแรงต่างๆ)

```python
set(data7['ov']) # ดูข้อมูลของคอลัมน์ ov
```

```python
data7['ov'] = pd.to_numeric(data7['ov'], errors='coerce') # errors='coerce' คือ ถ้าไม่เจอตัวเลขจะเปลี่ยนเป็นค่าว่าง
data7= data7.replace(np.nan, 0, regex=True) # เปลี่ยนจากค่าว่างให้เป็น 0
data7['ov'] = data7['ov'].astype(int) # เปลี่ยน type ของ ov ให้เป็น int
```

```python
set(data7['ov']) # ดูข้อมูลของคอลัมน์ ov อีกครั้ง
```

### PartitionKey 
เป็นตัวเลข 5 หลัก (ตัวเลขเท่านั้น)

```python
data7 = data7[(data7['PartitionKey'].astype(str).str.isdigit())& (data7['PartitionKey'].str.len() == 5)]    #.str.isdigit ใช้เพื่อตรวจสอบว่าค่าใดที่เป็นตัวเลข # เปลี่ยนคอลัมน์ user ให้มีตัวเลข 5 ตัว 
data7
```
