## ข้อมูลชุดที่ 2

```python
data3 = os.path.join(path,'alldb.csv') #แล้วก็ชี้ไปที่ไฟล์
print(data3)
```

```python
data4 = pd.read_csv(data3)  #กำหนดตัวแปร data4 เป็นข้อมูล ไฟล์ alldb.csv
data4  #เรียกดูตารางของข้อมูล ไฟล์ alldb.csv
```

#### ตัดเอาเฉพาะคอลัมน์ที่ต้องการ

```python
data5= data4[['user','RowKey','date','phone','site','time','urine','ov','PartitionKey']]
data5
```

#### ตรวจสอบว่าข้อมูลมีค่า Missing หรือไม่

```python
data5.isnull().any()  #ดูค่า Missing รายคอลัมน์ของตาราง data5
#เนื่องจากมี True แสดงว่ามีค่าที่เป็น Missing จะทำการลบค่า Missing ต่อไป
```

####แทน missing ด้วยค่าที่เหมาะสม 

โดยคำสั่ง fillna() ของตาราง data5

```python
data6 = data5.fillna(value={'user':'00000','RowKey':'ไม่ทราบ','date':'0000-00-00','phone':'ไม่ทราบ','site':'00000','time':'00-00-00','urine':'ไม่ทราบ','ov':'ไม่ทราบ','PartitionKey':'00000'}) #แทน missing ด้วยค่าที่เหมาะสม จากตาราง data5
data6 # แสดงตาราง data6
```
