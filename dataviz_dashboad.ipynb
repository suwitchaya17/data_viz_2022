## **Data Covid
ต้องการข้อมูลของวันที่ 1 ตุลาคม 2021 - 31 ธันวาคม 2021

## 2021-10-26 ถึง 2022-02-27

```python
data1 = os.path.join(path,'confirmed-cases-since-271064.xlsx') #แล้วก็ชี้ไปที่ไฟล์
print(data1)
```

```python
data2 = pd.read_excel(data1)  #กำหนดตัวแปร data_covid เป็นข้อมูลของผู้ป่วยโควิด
data2  #เรียกดูตารางของผู้ป่วยโควิด
```

## เลือกเอาเฉพาะข้อมูลวันที่ที่ต้องการ

```python
data3 = data2[(data2['announce_date']>='2021-10-01') & (data2['announce_date']<='2021-12-31')]
data3
```

#### ตัดเอาเฉพาะคอลัมน์ที่ต้องการ

```python
data4 = data3[['No.','announce_date','sex','age','risk','province_of_onset']]
data4
```

#### ตรวจสอบว่าข้อมูลมีค่า Missing หรือไม่

```python
data4.isnull().any()  #ดูค่า Missing รายคอลัมน์ของตาราง data4
#เนื่องจากมี True แสดงว่ามีค่าที่เป็น Missing จะทำการลบค่า Missing ต่อไป
```

####แทน missing ด้วยค่าที่เหมาะสม 

โดยคำสั่ง fillna() ของตาราง data_covid

```python
covid1_updated = data4.fillna(value={'No.':'ไม่ทราบ','announce_date':'ไม่ทราบ','sex':'ไม่ทราบ','age':'ไม่ทราบ','risk':'ไม่ทราบ'}) #แทน missing ด้วยค่าที่เหมาะสม จากตาราง data4
covid1_updated # แสดงตาราง covid1_updated
```

```python
covid_updated= covid1_updated.dropna() #ตัดค่าว่างออก
covid_updated
```

####ตรวจสอบว่ายังมีค่า Missing หรือไม่

```python
covid_updated.isnull().any() #ดูค่า Missing คอลัมน์ของตาราง covid_updated หลังแทน missing ด้วยค่าที่เหมาะสม
#เนื่องจากเป็น False ทุกค่า แสดงว่าตารางนี้ไม่มีค่า Missing แล้ว
```

##ตัดข้อมูลที่ไม่ทราบจังหวัดออก

```python
def nums(first_number, last_number, step=1): #ฟังก์ชั้นเรียกข้อมูลตามลำดับตั้งแต่เลขตัวแรกถึงตัวสุดท้ายโดยนับไปทีละ1
    return range(first_number, last_number+1, step) #ส่งข้อมูล
```

```python
from pandas.core.indexes.base import Index #นำเข้า index

frames = [] #สร้างตัวเก็บข้อมูล

Index_province = Index(data10['จังหวัด']) #นำรายชื่อจังหวัดออกมา
Index_province = Index_province.drop('ไม่ทราบ') #เอาข้อมูลที่ไม่ปรากฏชื่อจังหวัดออก
Index_province = Index_province.drop_duplicates(keep='first') #โดยเก็บชื่อจังหวัดที่พบเป็นอันดับแรกไว้ที่ตัวแปร index_province

for i in nums(0, 76): #สร้างลูปให้ i เป็นเลขตั้งแต่ 0-76 คือแต่ละจังหวัดที่ 0-76 ของตาราง
  mp_province = data10[data10['จังหวัด'] == Index_province[i]] # ให้ mp_province คือ ตัวแปรที่แยกข้อมูล77จังหวัด
  mp_province = mp_province.head(1) # แยก mp_province ให้เหลือ 5 ข้อมูลแรกของแต่ละจังหวัด
  frames.append(mp_province) #นำข้อมูล 1 ข้อมูลของแต่ละจังหวัดเข้าไปไว้ใน frame

mohpromt_updated = pd.concat(frames) # รวมข้อมูล
mohpromt_updated # แสดงผล
```

# ต่อตาราง data covid, data region และ mohpromt

```python
covid_region_mohpromt = covid_region_updated.merge(mohpromt_updated,how='left',left_on = 'จังหวัด',right_on = 'จังหวัด') 
#เชื่อมตาราง 2 ตาราง คือ covid_region_updated และ mohpromt_updated

covid_region_mohpromt #แสดงตาราง
```
